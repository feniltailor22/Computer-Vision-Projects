```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2
```

```python
img_stripes= cv2.imread('stripes.jpg')
img_stripes= cv2.cvtColor(img_stripes, cv2.COLOR_BGR2RGB)

pink_solid= cv2.imread('pink_solid.jpg')
pink_solid= cv2.cvtColor(pink_solid, cv2.COLOR_BGR2RGB)

f, (ax1, ax2)= plt.subplots(1,2)

ax1.imshow(img_stripes)
ax2.imshow(pink_solid)
```

```python
gray_stripes= cv2.cvtColor(img_stripes, cv2.COLOR_RGB2GRAY)
gray_solid= cv2.cvtColor(pink_solid, cv2.COLOR_RGB2GRAY)

f, (ax1, ax2)= plt.subplots(1,2)

ax1.imshow(gray_stripes)
ax2.imshow(gray_solid)
```

```python
#Normalizing the images

norm_stripes= gray_stripes/255.0
norm_solid= gray_solid/255.0
```

```python
#fourier transform function: it gives the output of high frequency changes in pixels in given image.

def ft_image(norm_img):
  f= np.fft.fft2(norm_img)
  fshift= np.fft.fftshift(f)
  frequency_tx= 20*np.log(np.abs(fshift))

  return frequency_tx
```

```python
f_stripes= ft_image(norm_stripes)
f_solid= ft_image(norm_solid)

f, (ax1, ax2, ax3, ax4)= plt.subplots(1,4,figsize=(20,10))

ax1.set_title('Original Stripes')
ax1.imshow(img_stripes)

ax2.set_title('fourier transformed stripes')
ax2.imshow(f_stripes, cmap='gray')

ax3.set_title('Pink Solid')
ax3.imshow(pink_solid)

ax4.set_title('fourier transformed pink solid')
ax4.imshow(f_solid, cmap='gray')
```

```python
birds_img= cv2.imread('birds.jpg')

birds_img= cv2.cvtColor(birds_img, cv2.COLOR_BGR2RGB)

gray_birds= cv2.cvtColor(birds_img, cv2.COLOR_RGB2GRAY)

f, (ax1, ax2)= plt.subplots(1,2, figsize=(15,10))

ax1.imshow(birds_img)
ax2.imshow(gray_birds)
```

```python
f_birds= ft_image(gray_birds)

plt.imshow(f_birds, cmap='gray')
```
