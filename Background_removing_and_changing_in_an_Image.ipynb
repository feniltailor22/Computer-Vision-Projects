```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2
```

```python
img= cv2.imread('pizza_bluescreen.jpg')
img= cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
img.shape
```

```python
lower_blue= np.array([0,0,250])

upper_blue= np.array([250,250,255])
```

```python
#Creating a mask
mask= cv2.inRange(img, lower_blue, upper_blue)

plt.imshow(mask, cmap='gray')
```

```python
plt.imshow(mask)  #without cmap
```

```python
#Removing background from the img
masked_img= np.copy(img)

masked_img[mask != 0]= [0,0,0]

plt.imshow(masked_img)
```

```python
bg_img= cv2.imread('space_background.jpg')
bg_img= cv2.cvtColor(bg_img, cv2.COLOR_BGR2RGB)

plt.imshow(bg_img)
```

```python
bg_img.shape, img.shape
```

```python
#Cropping bg_img to the img size
cropped_bg= bg_img[0:514, 0:816] 

plt.imshow(cropped_bg)
```

```python
cropped_bg.shape, img.shape
```

```python
#Cropped image with mask
cropped_bg[mask == 0] = [0,0,0]

plt.imshow(cropped_bg)
```

```python
final_img= masked_img + cropped_bg

plt.imshow(final_img)
```
