```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2
```

```python
img= mpimg.imread('birds.jpg')
plt.imshow(img)
```

```python
gray_img= cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray_img, cmap='gray')
```

```python
#printing gray pixel value at given [x,y] co-ordinates:
x=225
y=150

print(gray_img[x,y])
```

```python
#Maximum and minimum gray scale pixel value:
np.amax(gray_img),np.amin(gray_img)
```

```python
road_img= mpimg.imread('wa_state_highway.jpg')
```

```python
plt.imshow(road_img)
```

```python
#Isolating the RGB channels
r= road_img[:,:,0]
g= road_img[:,:,1]
b= road_img[:,:,2]
```

```python
r
```

```python
g
```

```python
b
```

```python
#Visualizing the individual color channels with cmap= 'gray'
```

```python
f, (ax1, ax2, ax3)= plt.subplots(1,3,figsize=(20,10))

ax1.set_title('R Channel')
ax1.imshow(r, cmap='gray')

ax2.set_title('G Channel')
ax2.imshow(g, cmap='gray')

ax3.set_title('B Channel')
ax3.imshow(b, cmap='gray')
```

```python
#Visualizing the individual color channels without cmap
```

```python
f, (ax1, ax2, ax3)= plt.subplots(1,3,figsize=(20,10))

ax1.set_title('R Channel')
ax1.imshow(r)

ax2.set_title('G Channel')
ax2.imshow(g)

ax3.set_title('B Channel')
ax3.imshow(b)
```
