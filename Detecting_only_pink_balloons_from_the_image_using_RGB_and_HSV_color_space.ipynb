```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2
```

```python
img= cv2.imread('water_balloons.jpg')
img= cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
r= img[:,:,0]
g= img[:,:,1]
b= img[:,:,2]
```

```python
f,(ax1, ax2, ax3)= plt.subplots(1,3,figsize=(20,10))

ax1.set_title('Red Channel')
ax1.imshow(r, cmap='gray')

ax2.set_title('Green Channel')
ax2.imshow(g, cmap='gray')

ax3.set_title('Blue Channel')
ax3.imshow(b, cmap='gray')

#Brighter pixel represents higher the value for the respective channels
```

```python
hsv_img= cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
#here we used RGB2HSV as img is already converted into RGB and now we want it in HSV color space.
plt.imshow(hsv_img)
```

```python
h= hsv_img[:,:,0]
s= hsv_img[:,:,1]
v= hsv_img[:,:,2]

f,(ax1, ax2, ax3)= plt.subplots(1,3,figsize=(20,10))

ax1.set_title('Hue Channel')
ax1.imshow(h, cmap='gray')

ax2.set_title('Saturation Channel')
ax2.imshow(s, cmap='gray')

ax3.set_title('Value Channel')
ax3.imshow(v, cmap='gray')
```

```python
#Setting the lower and upper bound mask (thresholding) for pink color

#RGB
lower_pink= np.array([180,0,100])
upper_pink= np.array([250,250,230])

#HSV
lower_hue= np.array([160,0,0])
upper_hue= np.array([180,255,255])
```

```python
#Masked Images

masked_rgb= cv2.inRange(img, lower_pink, upper_pink)

plt.imshow(masked_rgb)
```

```python
img[masked_rgb == 0]= [0,0,0]

plt.imshow(img)
```

```python
masked_hsv= cv2.inRange(hsv_img, lower_hue, upper_hue)

plt.imshow(masked_hsv)
```

```python
hsv_img[masked_hsv==0]= [0,0,0]

plt.imshow(hsv_img)
```
